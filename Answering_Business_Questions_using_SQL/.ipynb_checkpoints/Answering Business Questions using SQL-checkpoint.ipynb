```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def run_query(query):
    with sqlite3.connect('chinook.db') as conn:
        return pd.read_sql(query,conn)

def run_command(command):
    with sqlite3.connect('chinook.db') as conn:
        conn.isolation_level = None
        conn.execute(command)
```

```python
# Display all tables
def show_tables():
    query = '''
                SELECT
                    name,
                    type
                FROM sqlite_master
                WHERE type IN ("table","view");
            '''
    return run_query(query)

show_tables()
```

# Top Genre In USA

We need to help company to find out the best genre sold in the United States. Then they can focus stock on the most popular album. I will extract the information from database with multiple joins. The analysis will be using pandas and matplotlib.

Here is the list of top artist and genre of company consideration to get stock:
- Regal, Hip-Hop
- Red Tone, Punk
- Meteor and the Girls, Pop
- Slim Jim Bites, Blues

```python
top_genre_us_query = '''
        WITH usa_sales as
            (SELECT 
                    ar.name Artist_name,
                    g.name Genre,
                    SUM(il.quantity) Tracks_sold
             FROM customer c
             INNER JOIN invoice i ON i.customer_id = c.customer_id
             INNER JOIN invoice_line il ON il.invoice_id = i.invoice_id
             INNER JOIN track t ON t.track_id = il.track_id
             INNER JOIN genre g ON g.genre_id = t.genre_id
             INNER JOIN album al ON al.album_id = t.album_id
             INNER JOIN artist ar ON ar.artist_id = al.artist_id
             WHERE country = 'USA'
             GROUP BY Genre
            )
        SELECT
                Artist_name,
                Genre,
                Tracks_sold,
                CAST(Tracks_sold as float)/(SELECT SUM(tracks_sold) FROM usa_sales) Album_sales_ratio
        FROM usa_sales
        ORDER BY 3 DESC
        LIMIT 10;
        '''
top_genre_sales = run_query(top_genre_us_query)
top_genre_sales
```

```python
fig= plt.figure(figsize=(11,10))
ax = sns.barplot(x='Genre',y='Tracks_sold',data=top_genre_sales)
ax.set_xticklabels(top_genre_sales['Genre'],rotation=90)
ax.set_ylabel('Tracks sold')


# Display the sales ratio on the top of the bar chart
ratio_label = top_genre_sales['Album_sales_ratio']
for index in range(10):
    p = ax.patches[index]
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
           height + 3,
           '{:1.2f}'.format(ratio_label.iloc[index]*100)+'%',
           ha="center")
```

The bar chart is display the top genre sales in the United States. The summary with previously list is show below:
- Hip-Hop, 1.90% of album sold in the United States.
- Punk, 12.37% of album sold in the United States.
- Pop, 2.09% of album sold in the United States.
- Blues, 3.43% of album sold in the United States.

The summary is based on our dataset and I will recommend company to re-stock the genre of punk and blues.

# Employee Sales Performance

The company wants to see the performance of their employee. In this section, I will use SQL with multiple joins and subquery to calculate the total sales per employee. Also, data visualisation based on the data.

```python
employee_sales_performance_query =  '''
                                        WITH employee_sales_performance AS
                                            (SELECT
                                                    e.first_name || ' ' || e.last_name employee_name,
                                                    c.first_name || ' ' || c.last_name customer_name,
                                                    e.title employee_title,
                                                    e.hire_date,
                                                    SUM(i.total) total_sales
                                            FROM employee e
                                            INNER JOIN customer c ON c.support_rep_id = e.employee_id
                                            INNER JOIN invoice i ON i.customer_id = c.customer_id
                                            GROUP BY 1
                                            )
                                            
                                       SELECT
                                            employee_name,
                                            employee_title,
                                            hire_date,
                                            total_sales
                                       FROM employee_sales_performance
                                       GROUP BY 1;
                                    '''

employee_sales_performance = run_query(employee_sales_performance_query)
employee_sales_performance
```

```python
ax2 = sns.barplot(x='employee_name',y='total_sales',data=employee_sales_performance)
ax2.set_xlabel('')
```

The company is hired 3 **Sales Support Agents** at the moment. However, there are approximately 24% different between top sale (Jane Peacock) and bottom sale (Steve Johnson). Also, the amount of sales will be corresponded to the hire date.

# Analysing Sales by Country

The company wants to know what countries they need to focus. In this section, we still use SQL to obtain the average purchased by the customer and data visualisation with pie and bar chart.  

```python
sales_by_country_query = '''
                            WITH sales_by_country AS
                                (SELECT
                                        CASE
                                            WHEN(
                                                SELECT COUNT(*)
                                                FROM customer
                                                WHERE country = c.country
                                            ) = 1 THEN "Other"
                                            ELSE c.country
                                        END country,
                                        c.customer_id,
                                        il.*
                                  FROM invoice_line il
                                  INNER JOIN invoice i ON i.invoice_id = il.invoice_id
                                  INNER JOIN customer c ON c.customer_id = i.customer_id
                                )
                            
                            SELECT 
                                country,
                                count(distinct customer_id) customers_in_total,
                                SUM(unit_price) total_sales,
                                SUM(unit_price) / count(distinct invoice_id) average_order,
                                SUM(unit_price) / count(distinct customer_id) average_purchased_per_customer
                            FROM sales_by_country sbc
                            GROUP BY 1
                            ORDER BY CASE WHEN sbc.country = 'Other' THEN 1 ELSE 0 END, total_sales DESC;

'''
sales_by_country = run_query(sales_by_country_query)
sales_by_country
```

```python
#Create the plot and position
fig = plt.figure(figsize=(23,25))

#Bar chart for total of customers by country
ax1 = fig.add_subplot(2,2,1)
ax1 = sns.barplot(x='country',y='customers_in_total',data=sales_by_country)
ax1.set_xticklabels(sales_by_country['country'],rotation=90)
ax1.set_title('The total of customers by country')
ax1.set_xlabel('')

#Bar chart for total of sales by country
ax2 = fig.add_subplot(2,2,2)
ax2 = sns.barplot(x='country',y='total_sales',data=sales_by_country)
ax2.set_xticklabels(sales_by_country['country'],rotation=90)
ax2.set_xlabel('')
ax2.set_title('The total of sales by country')

#Pie chart for average order with percentage indicates
ax3 = fig.add_subplot(2,2,3)
ax3.pie(sales_by_country['average_order'],labels=sales_by_country['country'],autopct='%1.1f%%',
        shadow=False,pctdistance=0.8)
ax3.set_title('Percentage of average order')

#Pie chart for average purchased per customer with percentage indicate
ax4 = fig.add_subplot(2,2,4)
ax4.pie(sales_by_country['average_purchased_per_customer'],labels=sales_by_country['country'],
        autopct='%1.1f%%',shadow=False,pctdistance=0.9)
ax4.set_title('Percentage of Average Purchased Per Customer')
```

The Czech Republic have highest percentage of average order and average purchased per customer. On the other hand, the total of sales and customer is in the United States. However, I suggest the company to spend the money in the Czech Republic, because the Czech Republic has strong buying power potentially that shows from data.

# Album sales vs Track sales

The company is setup in a way that allows customer to make purchases in one of the two way:
- purchase a whole album
- purchase a collection of one or more individual tracks

The company now does not let customers purchase a whole album, and then add inidvidual tracks to that same purchase (unless they do that by choosing each track manually). When customers purchase albums they are charged the same price as if they had purchased each of those tracks separately.

Management are currently considering changing their purchasing strategy to save money. The strategy they are considering is to purchase only the most popular tracks from each album from record companies, instead of purchasing everytrack from an album.

```python
invoice_information ='''
                        WITH invoice_information AS
                            (SELECT 
                                t.track_id,
                                i.invoice_id,
                                i.total total
                             FROM track t
                             INNER JOIN invoice_line il ON il.track_id = t.track_id
                             INNER JOIN invoice i ON i.invoice_id = il.invoice_id
                             GROUP BY i.invoice_id
                            )
                        
                        SELECT
                            album_bought,
                            COUNT(*) invoices_in_total,
                            CAST(COUNT(*) as float)/(SELECT COUNT(invoice_id) FROM invoice_information) invoice_percentage,
                            SUM(total) total_sales,
                            CAST(SUM(total) as float)/(SELECT SUM(total) FROM invoice_information) sales_percentage
                        FROM
                            (SELECT
                                ivi.*,
                                CASE
                                    WHEN(
                                         SELECT t.track_id FROM track t
                                         WHERE t.album_id = (SELECT t2.album_id FROM track t2 WHERE t2.track_id=ivi.track_id)
                                         
                                         EXCEPT
                                         
                                         SELECT il2.track_id FROM invoice_line il2 WHERE il2.invoice_id=ivi.invoice_id
                                        ) IS NULL
                                    
                                    AND
                                        (SELECT il2.track_id FROM invoice_line il2 WHERE il2.invoice_id=ivi.invoice_id
                                        
                                        EXCEPT
                                        
                                        SELECT t.track_id FROM track t WHERE t.album_id=(SELECT t2.album_id FROM track t2 WHERE t2.track_id=ivi.track_id)
                                        ) IS NULL
                                    THEN 'YES'
                                    ELSE 'NO'
                                END album_bought
                            FROM invoice_information ivi
                            )
                        GROUP BY album_bought
'''
invoice_and_sales_compare = run_query(invoice_information)
invoice_and_sales_compare
```

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(1,2,1)
ax1.pie(invoice_and_sales_compare['invoice_percentage'],labels=['Track','Album'],autopct='%1.1f%%',
        shadow=False,pctdistance=0.8)
ax1.set_title('Invoice percentages between Track and Album')

ax2 = fig.add_subplot(1,2,2)
ax2.pie(invoice_and_sales_compare['sales_percentage'],labels=['Track','Album'],autopct='%1.1f%%',
        shadow=False,pctdistance=0.8)
ax2.set_title('Sales percentages between Track and Album')
```
